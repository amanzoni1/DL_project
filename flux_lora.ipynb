```python
import os

# Define the source folder containing original images
source_folder = '/content/drive/MyDrive/deep_learning/pics_example/CharAI'

# Define the destination folder for processed images and prompts
# You can name it as you prefer, e.g., 'CharAI_processed'
destination_folder = '/content/drive/MyDrive/deep_learning/pics_example/CharAI_processed'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)
```

```python
import shutil

# Supported image extensions
supported_extensions = ['.png', '.jpg', '.jpeg']

# List all files in the source folder
files = os.listdir(source_folder)

# Filter image files
image_files = [f for f in files if os.path.splitext(f)[1].lower() in supported_extensions]

# Sort the image files for consistent ordering
image_files.sort()

# Copy images to the destination folder
for filename in image_files:
    src_path = os.path.join(source_folder, filename)
    dst_path = os.path.join(destination_folder, filename)
    shutil.copy(src_path, dst_path)

print(f'Copied {len(image_files)} images to {destination_folder}')
```

```python
# Initialize counter
counter = 1

# List of files in the destination folder after copying
copied_files = os.listdir(destination_folder)

# Filter image files again to ensure consistency
copied_image_files = [f for f in copied_files if os.path.splitext(f)[1].lower() in supported_extensions]
copied_image_files.sort()

for filename in copied_image_files:
    # Get file extension
    ext = os.path.splitext(filename)[1].lower()

    # Define new filename
    new_filename = f'image_{counter}{ext}'

    # Define full paths
    old_path = os.path.join(destination_folder, filename)
    new_path = os.path.join(destination_folder, new_filename)

    # Rename the file
    os.rename(old_path, new_path)

    print(f'Renamed: {filename} --> {new_filename}')

    counter += 1
```

```python
# Necessary Imports
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
import torch
from PIL import Image
import matplotlib.pyplot as plt
import math
import os
```

```python
# Function to Initialize the Pipeline
def initialize_pipeline(pipe_id, models_folder, model_cp, lora_scale=1.0):
    # Load the pretrained pipeline
    pipeline = DiffusionPipeline.from_pretrained(
        pipe_id,
        torch_dtype=torch.float16
    )

    # Load LoRA weights
    pipeline.load_lora_weights(models_folder, weight_name=model_cp)

    # Move the pipeline to GPU
    pipeline = pipeline.to("cuda")

    # Optionally, you can set the scheduler if needed
    # For example, using DPMSolverMultistepScheduler for better performance
    #pipeline.scheduler = DPMSolverMultistepScheduler.from_config(pipeline.scheduler.config)

    return pipeline
```

```python
def generate_image_grid(pipeline, prompt, num_images=4, grid_size=(2, 2),
                       num_inference_steps=20, guidance_scale=4.0,
                       lora_scale=1.0, seed=42, save_images=False, save_folder=None):

    # Set the seed for reproducibility
    generator = torch.manual_seed(seed)

    # Generate images
    images = []
    for i in range(num_images):
        image = pipeline(
            prompt=prompt,
            num_inference_steps=num_inference_steps,
            guidance_scale=guidance_scale,
            #cross_attention_kwargs={"scale": lora_scale},
            generator=generator
        ).images[0]
        images.append(image)

    # Save images if required
    if save_images and save_folder:
        os.makedirs(save_folder, exist_ok=True)
        for idx, img in enumerate(images):
            img_path = os.path.join(save_folder, f"generated_image_{idx+1}.png")
            img.save(img_path)
            print(f"✅ Saved image: {img_path}")

    # Calculate grid size if not provided
    rows, cols = grid_size
    if rows * cols < num_images:
        cols = math.ceil(num_images / rows)

    # Create a matplotlib figure
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 4))

    # Flatten axes for easy iteration if grid_size > 1
    if isinstance(axes, Image.Image):
        axes = [axes]
    else:
        axes = axes.flatten()

    for idx, ax in enumerate(axes):
        if idx < num_images:
            ax.imshow(images[idx])
            ax.axis('off')
        else:
            ax.remove()  # Remove any unused subplots

    plt.tight_layout()
    plt.show()
```
