```python
import os

# Define the source folder containing original images
source_folder = '/content/drive/MyDrive/deep_learning/pics_example/CharAI'

# Define the destination folder for processed images and prompts
# You can name it as you prefer, e.g., 'CharAI_processed'
destination_folder = '/content/drive/MyDrive/deep_learning/pics_example/CharAI_processed'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)
```

```python
import shutil

# Supported image extensions
supported_extensions = ['.png', '.jpg', '.jpeg']

# List all files in the source folder
files = os.listdir(source_folder)

# Filter image files
image_files = [f for f in files if os.path.splitext(f)[1].lower() in supported_extensions]

# Sort the image files for consistent ordering
image_files.sort()

# Copy images to the destination folder
for filename in image_files:
    src_path = os.path.join(source_folder, filename)
    dst_path = os.path.join(destination_folder, filename)
    shutil.copy(src_path, dst_path)

print(f'Copied {len(image_files)} images to {destination_folder}')
```

```python
# Initialize counter
counter = 1

# List of files in the destination folder after copying
copied_files = os.listdir(destination_folder)

# Filter image files again to ensure consistency
copied_image_files = [f for f in copied_files if os.path.splitext(f)[1].lower() in supported_extensions]
copied_image_files.sort()

for filename in copied_image_files:
    # Get file extension
    ext = os.path.splitext(filename)[1].lower()

    # Define new filename
    new_filename = f'image_{counter}{ext}'

    # Define full paths
    old_path = os.path.join(destination_folder, filename)
    new_path = os.path.join(destination_folder, new_filename)

    # Rename the file
    os.rename(old_path, new_path)

    print(f'Renamed: {filename} --> {new_filename}')

    counter += 1
```

```python
import re

def natural_sort(l):
    """Sort the given list in the way that humans expect."""
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ]
    return sorted(l, key=alphanum_key)

# Supported image extensions
supported_extensions = ['.png', '.jpg', '.jpeg']

# List all image files in the destination folder
all_files = os.listdir(destination_folder)
image_files = [f for f in all_files if os.path.splitext(f)[1].lower() in supported_extensions]

# Sort the image files naturally
sorted_image_files = natural_sort(image_files)

print(f"Total images found: {len(sorted_image_files)}")
print("List of sorted images:")
for img in sorted_image_files:
    print(img)
```

```python
# Proceed only if the number of prompts matches the number of images
if num_prompts == num_images:
    for img, prompt in zip(sorted_image_files, prompts):
        # Extract the base name without extension
        base_name = os.path.splitext(img)[0]

        # Define the prompt filename
        txt_filename = f'{base_name}.txt'

        # Define the full path for the prompt file
        txt_path = os.path.join(destination_folder, txt_filename)

        # Write the prompt to the .txt file
        with open(txt_path, 'w') as f:
            f.write(prompt)

        print(f'✅ Created prompt file: {txt_filename}')
else:
    print("❌ Cannot proceed with prompt assignment due to mismatched counts.")
```
